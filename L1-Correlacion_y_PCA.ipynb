```python
#Cargamos librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
#descargamos los datos
data = datasets.load_wine()
```

```python
data
```

```python
# Creamos un DataFrame con los datos, los nombres de las columnas serán los feature_names
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df
```

```python
# Resumen estadístico de los datos
df.describe()
```

```python
# Calcular la matriz de correlaciones llamamos directamente a la función corr
corr = df.corr()
```

```python
corr
```

```python
# Para visualizar la matriz de correlaciones, usamos la función heatmap
sns.heatmap(corr)
```

```python
# Un mapa de colores lineal no es la mejor opción para visualizar una variable que va de valores negativos a positivos
sns.heatmap(corr, cmap='coolwarm_r')
```

```python
# Por defecto el mapa de colores va desde el valor mínimo de los datos hasta el máximo. Si queremos que la escala quede centrada en 0, debemos especificar el rango
sns.heatmap(corr, cmap='coolwarm_r', vmin=-1., vmax=1.)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.heatmap(corr, cmap='coolwarm_r', vmin=-1., vmax=1., ax=ax)
ax.set_title('Pearson Correlation')
```

```python
# Guardamos
fig.savefig('pearson.png')
```

```python
# Y finalmente la correlación de kendall
fig, ax = plt.subplots(figsize=(6,4))
corr3 = df.corr(method='kendall')
sns_plot = sns.heatmap(corr3, cmap='coolwarm_r', square=True, vmin=-1, vmax=1)
ax.set_title('Kendall correlation')
```

```python
# Ahora pasaremos al Analisis de Componente Principal. Primero importamos las librerias necesarias
from sklearn.decomposition import PCA # función para calcular PCA
from sklearn.pipeline import make_pipeline # función para crear flujos de trabajo
from sklearn.preprocessing import StandardScaler # funciones para re-escalar los datos
from sklearn.preprocessing import scale
```

```python
# Creamos un flujo que re-escalará los datos, y luego calculará PCA
pca_pipe = make_pipeline(StandardScaler(), PCA())
pca_pipe.fit(df) # calculamos el PCA
```

```python
# Creamos una variable para analizar los resultados del PCA
model_pca = pca_pipe.named_steps['pca']
```

```python
model_pca.components_
```

```python
# Creamos un DataFrame a partir de los resultados
dat = pd.DataFrame(data=model_pca.components_,
                   columns=df.columns,
                   index=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13'])
```

```python
dat
```

```python
# Para analizar los resultados, visualizaremos la matriz de pesos.
fig, ax = plt.subplots(figsize=(6,5))
components = model_pca.components_
max_value = max(components.max(),abs(components.min()))
plt.imshow(components.T, cmap='coolwarm', vmin=-max_value, vmax=max_value)
plt.yticks(range(len(df.columns)), df.columns)
plt.xticks(range(len(df.columns)), dat.index)
plt.colorbar()
plt.title('PCA components')
```

```python
# Para ver qué tan bien estos componentes principales capturan los datos, analizaremos la fracción de varianza acumulada
prop_variance_acum = model_pca.explained_variance_ratio_.cumsum()
print(prop_variance_acum)
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(np.arange(len(df.columns))+1, prop_variance_acum, marker='o')
for x, y in zip(np.arange(len(df.columns))+1, prop_variance_acum):
  label = round(y,2)
  ax.annotate(label, (x,y), xytext=(x, y+.1), ha='center')

ax.set_xticks(np.arange(model_pca.n_components_)+1)
ax.set_xticklabels(dat.index)
ax.set_ylim([0,1.3])
ax.set_title(('Explained variance'))
ax.set_xlabel('Principal Component')
ax.set_ylabel('% accumulated variance')
```
